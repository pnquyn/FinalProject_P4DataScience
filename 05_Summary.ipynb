**"Sự thật về Khách hàng quay lại"**

Điều bất ngờ nhất trong toàn bộ quá trình phân tích là việc các mô hình học máy đã đánh giá tiêu cực đối với biến số Khách hàng quay lại. (Mô hình `Random Forest` đánh trọng số âm cho biến `Returning_Visitor` (~ -0.8))

Thông thường, chúng ta mặc định rằng khách hàng cũ là những người trung thành và dễ mua hàng nhất. Tuy nhiên, dữ liệu thực tế lại chứng minh điều ngược lại: việc một người dùng được định danh là "Khách cũ" lại là dấu hiệu cho thấy xác suất mua hàng thấp đi. Họ truy cập nhiều nhưng thoát nhanh và phần lớn được phân loại vào nhóm "Vãng lai" hoặc "So sánh giá".

Phát hiện này là một sự phản biện mạnh mẽ đối với các giả định cảm tính. Nó chứng minh rằng nếu không có các ưu đãi cá nhân hóa đặc biệt, tệp khách hàng cũ có thể trở thành gánh nặng cho hệ thống vận hành thay vì là nguồn doanh thu như kỳ vọng.

**1. Câu hỏi bổ sung**
* **_Dự đoán giá trị đơn hàng:_** Mở rộng bài toán từ Phân loại (`Classification`) sang Hồi quy (`Regression`): Hiện tại mô hình chỉ dự đoán được việc mua hay không. Hướng phát triển tiếp theo là thu thập thêm dữ liệu về giá trị giỏ hàng thực tế để xây dựng mô hình dự báo, giúp phân định rõ phân khúc khách hàng VIP. 
* **_Tại sao khách hàng cũ lại chốt đơn thấp?_** Dữ liệu cho thấy `Returning Visitor` có tỷ lệ chuyển đổi thấp. Cần tích hợp thêm Dữ liệu `Clickstream` (Luồng click) để xác định xem họ quay lại để so sánh giá, kiểm tra tình trạng đơn hàng hay chỉ dạo chơi. Từ đó đề xuất chiến lược `Retargeting` (quảng cáo bám đuổi) phù hợp.
* **_Website có bị lỗi trên trình duyệt hoặc hệ điều hành cụ thể nào không?_** Phân tích chỉ số `Bounce Rate` và `Exit Rate` trên từng cặp `Browser/OS` cụ thể. Mục tiêu là phát hiện các lỗi hiển thị hoặc lỗi hiệu năng khiến người dùng thoát trang đột ngột trên các thiết bị cụ thể.

**2. Thực hiện phân tích sâu hơn**
* **_Phân tích Độ nhạy (Sensitivity Analysis):_** Thực hiện các bài test "What-if" bằng cách thay đổi các biến số đầu vào của mô hình. Trả lời được câu hỏi chiến lược "Nếu giảm `ExitRate` đi 5% thì doanh thu dự kiến tăng bao nhiêu?". Điều này giúp định lượng được giá trị của việc cải thiện UX/UI, từ đó xếp hạng ưu tiên cho các đầu việc kỹ thuật.
* **_Tìm "Điểm gãy" (Drop-off Analysis):_** Xác định ngưỡng kiên nhẫn của người dùng. Trả lời được câu hỏi "Khách hàng thường bỏ cuộc sau khi xem bao nhiêu trang sản phẩm?". Nếu ngưỡng này thấp, cần tối ưu lại công cụ tìm kiếm và bộ lọc sản phẩm để khách tìm thấy đồ nhanh hơn.
* **_Tâm lý mùa vụ (Seasonality Impact):_** Phân tích sâu sự thay đổi hành vi vào các dịp lễ. Xem xét liệu vào ngày lễ, khách hàng có xu hướng quyết định nhanh hơn (mua theo cảm xúc) hay so sánh kỹ hơn so với ngày thường để điều chỉnh thông điệp quảng cáo.

**3. Cải tiến kỹ thuật mô hình**
* **_Nâng cấp thuật toán:_** Chuyển từ `Random Forest` sang `XGBoost` hoặc `LightGBM`. Các thuật toán `Boosting` này xử lý tốt hơn các dữ liệu dạng bảng và các mối quan hệ phi tuyến tính phức tạp, giúp tăng độ chính xác.
* **_Xử lý mất cân bằng nâng cao:_** Áp dụng kỹ thuật `SMOTE` (sinh dữ liệu giả lập). Thay vì chỉ điều chỉnh trọng số, `SMOTE` tạo ra các mẫu giả lập cho nhóm khách "Mua", giúp mô hình học được đặc điểm của nhóm thiểu số tốt hơn.
* **_Tối ưu hóa tự động:_** Sử dụng `GridSearchCV` hoặc `Bayesian Optimization`. Thay thế việc chọn tham số thủ công/mặc định bằng cách tìm kiếm bộ tham số toán học tối ưu nhất để cực đại hóa `F1-score`.

**4. Mở rộng dữ liệu**
* **_Dữ liệu Doanh thu:_** Chuyển bài toán từ phân loại "Mua/Không" sang hồi quy "Dự đoán doanh thu", giúp xác định và chăm sóc nhóm khách hàng VIP (High-value users).
* **_Định danh người dùng:_** Liên kết các phiên truy cập rời rạc của cùng một người để tính toán Giá trị vòng đời khách hàng và tỷ lệ quay lại thực tế.
* **_Ngữ cảnh sản phẩm:_** Bổ sung thông tin danh mục (Thời trang, Gia dụng...) để hiểu sâu sở thích người dùng, phục vụ cho hệ thống gợi ý.

**5. Ứng dụng thực tế**
* **_Hệ thống kích cầu thời gian thực:_** Tích hợp mô hình vào website. Khi khách đạt điểm tiềm năng (`PageValues`) cao, hệ thống tự động hiện Mã giảm giá (`Coupon`) để "chốt đơn" ngay lập tức.
* **_Cá nhân hóa trải nghiệm:_** Tùy biến giao diện dựa trên hành vi. Ví dụ: Nhóm khách thích tìm hiểu (`Informational`) sẽ được ưu tiên hiển thị các bài Blog/Review sản phẩm trên trang chủ.
* **_Kiểm chứng hiệu quả:_** Chạy thử nghiệm trên nhóm nhỏ khách hàng thực tế để đo lường chính xác xem việc áp dụng mô hình giúp tăng bao nhiêu % doanh thu so với cách cũ.

##### **Những trở ngại nào đã xuất hiện trong quá trình thực hiện?**

**_Hân_**
* Dữ liệu bị mất cân bằng nghiêm trọng. Biến `PageValues` có giá trị 0 chiếm số đông, trong khi các giá trị cao lại rất thưa thớt. Biến `VisitorType` thì nhóm Khách cũ áp đảo hoàn toàn Khách mới.
* Một số biến phân loại có độ đa dạng quá cao như `TrafficType` (20 loại) hay `Browser` (13 loại), gây khó khăn khi trực quan hóa và tìm ra quy luật chung.
* Dữ liệu bị khuyết thiếu về mặt thời gian (không có dữ liệu tháng 1 và tháng 4), khiến việc phân tích xu hướng theo mùa vụ có thể bị đứt gãy.

***Thư:***

- Vì kiến thức nghiệp vụ về lĩnh vực này còn hạn chế, nên khi trực quan và phân tích các số liệu, lúc đầu chỉ có thể phân tích mô tả mà không biết insight thực sự ẩn sâu bên trong là gì.
- Khi có được thông tin trên dữ liệu, khó khăn lớn nhất là có thể đặt ra câu hỏi nghiên cứu, để câu hỏi đó có thể mang lại giá trị thực sự.
- Đưa ra hướng tiền xử lý để huấn luyện model là một công việc khá khó khăn. Vì phải hiểu rõ về dữ liệu và hiểu rõ một phần nguyên nhân tại sao dữ liệu lại như vậy. Sau đó mới cân nhắc chọn ra đặc trưng nào, có nên bỏ đặc trưng nào không, với mỗi đặc trưng nhất định thì nên biến đổi như thế nào, nên tạo ra đặc trưng nào mới. Vì mỗi lựa chòn đều ảnh hưởng trực tiếp đến việc huấn luyện mô hình.

***Quỳnh:***

- Do đây là dự án phân tích sâu đầu tiên trên tập dữ liệu lớn, việc xác định các câu hỏi nghiên cứu sao cho hợp lý và mang lại ý nghĩa thực tế gặp nhiều lúng túng trong giai đoạn khởi đầu.

- Quá trình vẽ biểu đồ gặp khó khăn trong việc truyền tải thông tin hiệu quả. Các biểu đồ ban đầu thường bị nhồi nhét quá nhiều dữ liệu, gây rối mắt và khó nắm bắt được thông điệp chính.

- Xung đột mã nguồn (merge code) xảy ra thường xuyên do cơ chế lưu trữ đặc thù của file notebook, dẫn đến nhiều lỗi logic khi tích hợp phần làm việc của các thành viên.

##### **Đã làm thế nào để khắc phục các vấn đề đó?**

**_Hân_**
* Sử dụng kỹ thuật Chia nhóm (Binning). Thay vì để `PageValues` là biến liên tục, nhóm đã chia thành các khoảng (0, 1-10, 11-20...) để nhìn rõ xu hướng chuyển đổi tuyến tính.
* Áp dụng các Kiểm định thống kê để xác nhận sự khác biệt có ý nghĩa hay không, thay vì chỉ dựa vào cảm quan biểu đồ.
* Tập trung phân tích sâu vào Tỷ lệ chuyển đổi (`Conversion Rate`) và Hành vi (`Bounce Rate`) thay vì chỉ nhìn vào số lượng truy cập thuần túy để đánh giá chất lượng thực sự của từng nhóm.

***Thư***

- Tra cứu về các thuật ngữ được sử dụng trên tập dữ liệu để nắm bắt được cơ vản kiến thức nghiệp vụ của mà dữ liệu thu thập. Tự mình insight, trao đổi với các thành viên và thử insight bằng AI.  Áp dụng phương pháp **đối chứng chéo**, đối chiếu trên dữ liệu để biết được rẳng insight của mình có đảm bảo tính khách quan và chính xác hay không.
  
- Tự đặt bản thân mình vào vị trí của **Stackholders**. Tự hỏi bản thân: *Nếu là chủ sỡ hữu tập dữ liệu này, thì cần câu trả lời nào để tăng doanh thu.* Hiểu rõ các cột đặc trưng, với mỗi đặc trưng thì để trả lời cho câu hỏi phân tích đặc trưng này để làm gì, tại sao người ta cần thu thập đặc trưng gì, các ý nghĩa có thể suy ra từ đặc trưng này. Sau đó, xem thử trên tập dữ liệu, có thông tin nào có thể đóng góp để trả lời được thắc mắc của mình không, cuối cùng là kết hợp các thông tin lại để có thể đặt ra câu hỏi hoàn chỉnh.

  
- Về phần này, trước khi giải quyết thì đọc thêm về các kĩ thuật xử lý như kĩ thuật Scale, các kĩ thuật Feature Engineering, tham khảo cách những người đi trước khi nào sử dụng nó, sử dụng nó để làm gì, không sử dụng nó thì kết quả có nguy cơ bị ảnh hưởng như thế nào. Đồng thời, khi biết được mình lựa chọn mô hình gì, thì các biến của mình phải được mã hóa hay Scale ra sao rồi lựa chọn cho phụ hợp.

***Quỳnh:***

- Tham khảo các bài phân tích mẫu trên Kaggle để học hỏi tư duy tiếp cận, khám phá lại dữ liệu để đặt ra câu hỏi hợp lí.

- Tinh chỉnh lại biểu đồ theo gợi ý của giáo viên hướng dẫn

- Chia các file riêng để hạn chế việc phần làm của các thành viên bị chồng lên nhau

##### **Đâu là thử thách lớn nhất và nguyên nhân là gì?**

**_Hân:_** Dữ liệu cho thấy khách cũ truy cập rất nhiều nhưng tỷ lệ mua thấp hơn khách mới. Thử thách là phải tìm ra lý do nên phải đào sâu vào các chỉ số phụ như `BounceRates` và `ExitRates` mới kết luận được đây là hành vi "window shopping" bình thường.

**Thư** 
- Thử thách lớn nhất rơi vào hướng tiền xử lý dữ liệu trước khi huấn luyện model. Bởi vì với những thay đổi nhỏ và cách xử lý trên từng đặc trưng thì mô hình sẽ bị ảnh hưởng. Những lần huấn luyện model cho ra kết quả không tốt, thì mới nhận ra được bước tiền xử lý của mình đã thiếu sót hoặc làm đánh mất đặc trưng quan trọng, hoặc không biết cách biến đổi đặc trưng để mô hình học được tốt trên dữ liệu. Đôi khi còn gặp khó khăn trong việc xác định nguyên nhân khiến mô hình bị sai, không rõ là do từ cách mình xử lý hay đến từ thông tin của dữ liệu. Vì thế nên quá trình này bản thân phải thử và sai nhiều lần, qua một lần lại rút ra được mình bị sai sót chỗ nào và điều chỉnh lại. 

***Quỳnh***: Giai đoạn đặt câu hỏi được xem là thách thức lớn nhất. Trong khi các kỹ thuật lập trình có thể tra cứu nhanh, việc tư duy để đặt các vấn đề kinh doanh cần giải quyết đòi hỏi nhiều thời gian nghiên cứu và hiểu về dữ liệu.

##### **Những kiến thức và kỹ năng nào đã được tích lũy?**

**_Hân_**
* **_Kỹ năng Xử lý dữ liệu:_** Biết cách chuẩn hóa dữ liệu (Standardization) và xử lý các giá trị ngoại lai (outliers) một cách hợp lý.
* **_Kỹ năng Thống kê:_** Hiểu và vận dụng được các kiểm định phi tham số khi dữ liệu không tuân theo phân phối chuẩn.
* **_Tư duy Kinh doanh:_** Học cách chuyển đổi các con số thống kê thành Insight hành động.

***Thư***

- Biết được quy trình chuẩn chỉnh từ xử lý dữ liệu đến đưa ra những insight mang lợi ích thực sự.

- Có thêm nhiều kinh nghiệm trong việc xử lý đặc trưng trong từng trường hợp  trước bước huấn luyện model để model có thể nắm bắt được thông tin từ các đặc trưng tối ưu.

- Biết được cách đặt vấn đề trên dữ liệu và từ chính dữ liệu đi phân tích để giải quyết vấn đề.

***Quỳnh:***

- Hiểu được rằng giá trị cốt lõi của trực quan hóa nằm ở sự rõ ràng, súc tích và dễ hiểu thay vì màu sắc hay kỹ thuật phức tạp.

- Hình thành được kỹ năng chuyển đổi từ dữ liệu sang các câu hỏi thực tế thông qua việc tham khảo và học hỏi.

- Rút ra được kinh nghiệm quan trọng trong việc sắp xếp cấu trúc file và tổ chức mã nguồn để quy trình làm việc nhóm diễn ra khoa học, trơn tru hơn.

##### **Điều gì gây bất ngờ nhất trong quá trình thực hiện?**

**_Hân_**
* Về `PageValues`: Chỉ cần chỉ số này nhích nhẹ từ 0 lên khoảng 0-10, tỷ lệ chốt đơn đã tăng vọt gấp 10 lần. Đây là "điểm gãy" quan trọng nhất để phân loại khách hàng.
* Về `VisitorType`: Trái ngược với suy nghĩ "khách quen là nhất", dữ liệu chỉ ra Khách hàng mới mới là nhóm "chịu chi" hơn với tỷ lệ chuyển đổi ~25% (so với 14% của khách cũ).

***Thư:***  Điểm gây bất ngờ lớn nhất đó chính là sức mạnh của giai đoạn **Tiền xử lý dữ liệu (Data Preprocessing)**. Vì lần đầu khi huấn luyện mô hình, bản thân chỉ áp dụng một cách máy móc các phương pháp xử lý tham khảo từ các nguồn bên ngoài mà chưa thực sự đặt câu hỏi "tại sao" hoặc xem xét liệu nó có phù hợp với đặc thù của tập dữ liệu hiện tại hay không. Sau khi mô hình đạt được giá trị rất thấp, mới thử tiền xử lý dữ liệu lại và nhận được kết quả tốt hơn ở mỗi lần. Từ đó mới thấy số lượng dữ liệu lớn không quan trọng bằng việc chúng ta **"hiểu"** dữ liệu đó đến mức nào. Khai thác dữ liệu tối ưu không nằm ở thuật toán phức tạp, mà nằm ở khả năng biến đổi và xử lý dữ liệu sao cho mô hình có thể tiếp nhận thông tin hiệu quả nhất.


***Quỳnh***: Điểm bất ngờ lớn nhất là giai đoạn chuẩn bị (đặt câu hỏi, làm sạch dữ liệu) chiếm phần lớn thời gian và quyết định kết quả, thay vì giai đoạn chạy mô hình hay viết thuật toán như những suy nghĩ ban đầu.

##### **Dự án đã định hình lại tư duy về Khoa học dữ liệu như thế nào?**

**_Hân_**
* Nhận ra rằng làm sạch và Tiền xử lý mới là giai đoạn quan trọng nhất, chiếm 70-80% thời gian và quyết định độ chính xác của mô hình, chứ không phải là việc chọn thuật toán phức tạp.
* Khoa học dữ liệu không chỉ là Code, mà là Kể chuyện (`Storytelling`). Kết quả phân tích chỉ có giá trị khi nó được diễn giải thành ngôn ngữ mà bộ phận kinh doanh/marketing có thể hiểu và dùng được.
* Quá trình làm việc không phải là đường thẳng mà là vòng lặp "Thử nghiệm - Sai - Sửa". Không có giả định nào là đúng tuyệt đối cho đến khi được dữ liệu kiểm chứng.

***Thư****: 
- Lúc đầu, bản thân có cái nhìn khá đơn giản về Khoa học dữ liệu, vì nghĩ công việc chỉ đơn giản là phân tích thông tin để đưa ra các nhận xét. Khi được học lý thuyết, bản thân chưa hình dung ra được những cái khó ở mỗi giai đoạn mà thầy cô hay nhắc đến.

- Sau khi hoàn thành dự án thì bản thân nhận ra Khoa học dữ liệu đòi hỏi ở một người không chỉ là kiến thức chuyên ngành, mà phải nắm bắt được kiến thức nghiệp vụ (domain knowledge), tính cẩn thận và kiên nhẫn. Nếu không có kiến thức nghiệp vụ, những con số chỉ là những con số và không có insight nào được sinh ra, cũng sẽ không có bài toán nào được đặt ra và dữ liệu trở nên vô nghĩa. Mọi bước trong quy trình, từ tiền xử lý dữ liệu đến lựa chọn đặc trưng, đều đòi hỏi tư duy phản biện và sự suy xét kỹ lưỡng. Một quyết định sai lầm ở bước xử lý dữ liệu có thể dẫn đến sự sai lệch hoàn toàn ở giai đoạn cuối. cần sự kiên nhẫn khi thấy sai và quay lại sửa sai. Vì thế mà tính kiên nhẫn và sự tỉ mỉ luôn được để cao, để tránh sai sót ít nhất có thể. Đồng thời, sẵn sàng đối mặt với chuyện bị sai và quay lại sửa sai.

***Quỳnh***: Dự án giúp nhận ra Khoa học dữ liệu là một quá trình "Thử nghiệm - Sai - Sửa" liên tục. Không có giải pháp nào hoàn hảo ngay từ bước đầu tiên, mà cần sự kiên nhẫn tinh chỉnh từng bước nhỏ, từ cách đặt câu hỏi định hướng cho đến cách trình bày kết quả cuối cùng.
