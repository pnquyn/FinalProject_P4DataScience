##### **Những trở ngại nào đã xuất hiện trong quá trình thực hiện?**

Hân

***Thư:***

- Vì kiến thức nghiệp vụ về lĩnh vực này còn hạn chế, nên khi trực quan và phân tích các số liệu, lúc đầu chỉ có thể phân tích mô tả mà không biết insight thực sự ẩn sâu bên trong là gì.
- Khi có được thông tin trên dữ liệu, khó khăn lớn nhất là có thể đặt ra câu hỏi nghiên cứu, để câu hỏi đó có thể mang lại giá trị thực sự.
- Đưa ra hướng tiền xử lý để huấn luyện model là một công việc khá khó khăn. Vì phải hiểu rõ về dữ liệu và hiểu rõ một phần nguyên nhân tại sao dữ liệu lại như vậy. Sau đó mới cân nhắc chọn ra đặc trưng nào, có nên bỏ đặc trưng nào không, với mỗi đặc trưng nhất định thì nên biến đổi như thế nào, nên tạo ra đặc trưng nào mới. Vì mỗi lựa chòn đều ảnh hưởng trực tiếp đến việc huấn luyện mô hình.

***Quỳnh:***

- Do đây là dự án phân tích sâu đầu tiên trên tập dữ liệu lớn, việc xác định các câu hỏi nghiên cứu sao cho hợp lý và mang lại ý nghĩa thực tế gặp nhiều lúng túng trong giai đoạn khởi đầu.

- Quá trình vẽ biểu đồ gặp khó khăn trong việc truyền tải thông tin hiệu quả. Các biểu đồ ban đầu thường bị nhồi nhét quá nhiều dữ liệu, gây rối mắt và khó nắm bắt được thông điệp chính.

- Xung đột mã nguồn (merge code) xảy ra thường xuyên do cơ chế lưu trữ đặc thù của file notebook, dẫn đến nhiều lỗi logic khi tích hợp phần làm việc của các thành viên.

##### **Đã làm thế nào để khắc phục các vấn đề đó?**

Hân

***Thư***

- Tra cứu về các thuật ngữ được sử dụng trên tập dữ liệu để nắm bắt được cơ vản kiến thức nghiệp vụ của mà dữ liệu thu thập. Tự mình insight, trao đổi với các thành viên và thử insight bằng AI.  Áp dụng phương pháp **đối chứng chéo**, đối chiếu trên dữ liệu để biết được rẳng insight của mình có đảm bảo tính khách quan và chính xác hay không.
  
- Tự đặt bản thân mình vào vị trí của **Stackholders**. Tự hỏi bản thân: *Nếu là chủ sỡ hữu tập dữ liệu này, thì cần câu trả lời nào để tăng doanh thu.* Hiểu rõ các cột đặc trưng, với mỗi đặc trưng thì để trả lời cho câu hỏi phân tích đặc trưng này để làm gì, tại sao người ta cần thu thập đặc trưng gì, các ý nghĩa có thể suy ra từ đặc trưng này. Sau đó, xem thử trên tập dữ liệu, có thông tin nào có thể đóng góp để trả lời được thắc mắc của mình không, cuối cùng là kết hợp các thông tin lại để có thể đặt ra câu hỏi hoàn chỉnh.

  
- Về phần này, trước khi giải quyết thì đọc thêm về các kĩ thuật xử lý như kĩ thuật Scale, các kĩ thuật Feature Engineering, tham khảo cách những người đi trước khi nào sử dụng nó, sử dụng nó để làm gì, không sử dụng nó thì kết quả có nguy cơ bị ảnh hưởng như thế nào. Đồng thời, khi biết được mình lựa chọn mô hình gì, thì các biến của mình phải được mã hóa hay Scale ra sao rồi lựa chọn cho phụ hợp.

***Quỳnh:***

- Tham khảo các bài phân tích mẫu trên Kaggle để học hỏi tư duy tiếp cận, khám phá lại dữ liệu để đặt ra câu hỏi hợp lí.

Hân

**Thư**

Những lần huấn luyện model cho ra kết quả không tốt, thì mới nhận ra được bước tiền xử lý của mình đã thiếu sót hoặc làm đánh mất đặc trưng quan trọng, hoặc không biết cách biến đổi đặc trưng để mô hình học được tốt trên dữ liệu. ***Quỳnh***: Giai đoạn đặt câu hỏi được xem là thách thức lớn nhất. Trong khi các kỹ thuật lập trình có thể tra cứu nhanh, việc tư duy để đặt các vấn đề kinh doanh cần giải quyết đòi hỏi nhiều thời gian nghiên cứu và hiểu về dữ liệu.

***Thư:***  Điểm gây bất ngờ lớn nhất đó chính là sức mạnh của giai đoạn **Tiền xử lý dữ liệu (Data Preprocessing)**. Vì lần đầu khi huấn luyện mô hình, bản thân chỉ áp dụng một cách máy móc các phương pháp xử lý tham khảo từ các nguồn bên ngoài mà chưa thực sự đặt câu hỏi "tại sao" hoặc xem xét liệu nó có phù hợp với đặc thù của tập dữ liệu hiện tại hay không. Sau khi mô hình đạt được giá trị rất thấp, mới thử tiền xử lý dữ liệu lại và nhận được kết quả tốt hơn ở mỗi lần. Từ đó mới thấy số lượng dữ liệu lớn không quan trọng bằng việc chúng ta **"hiểu"** dữ liệu đó đến mức nào. Khai thác dữ liệu tối ưu không nằm ở thuật toán phức tạp, mà nằm ở khả năng biến đổi và xử lý dữ liệu sao cho mô hình có thể tiếp nhận thông tin hiệu quả nhất.


***Quỳnh***: Điểm bất ngờ lớn nhất là giai đoạn chuẩn bị (đặt câu hỏi, làm sạch dữ liệu) chiếm phần lớn thời gian và quyết định kết quả, thay vì giai đoạn chạy mô hình hay viết thuật toán như những suy nghĩ ban đầu.

##### **Dự án đã định hình lại tư duy về Khoa học dữ liệu như thế nào?**

Hân

***Thư****: 
- Lúc đầu, bản thân có cái nhìn khá đơn giản về Khoa học dữ liệu, vì nghĩ công việc chỉ đơn giản là phân tích thông tin để đưa ra các nhận xét. Khi được học lý thuyết, bản thân chưa hình dung ra được những cái khó ở mỗi giai đoạn mà thầy cô hay nhắc đến.

- Sau khi hoàn thành dự án thì bản thân nhận ra Khoa học dữ liệu đòi hỏi ở một người không chỉ là kiến thức chuyên ngành, mà phải nắm bắt được kiến thức nghiệp vụ (domain knowledge), tính cẩn thận và kiên nhẫn. Nếu không có kiến thức nghiệp vụ, những con số chỉ là những con số và không có insight nào được sinh ra, cũng sẽ không có bài toán nào được đặt ra và dữ liệu trở nên vô nghĩa. Mọi bước trong quy trình, từ tiền xử lý dữ liệu đến lựa chọn đặc trưng, đều đòi hỏi tư duy phản biện và sự suy xét kỹ lưỡng. Một quyết định sai lầm ở bước xử lý dữ liệu có thể dẫn đến sự sai lệch hoàn toàn ở giai đoạn cuối. cần sự kiên nhẫn khi thấy sai và quay lại sửa sai. Vì thế mà tính kiên nhẫn và sự tỉ mỉ luôn được để cao, để tránh sai sót ít nhất có thể. Đồng thời, sẵn sàng đối mặt với chuyện bị sai và quay lại sửa sai.

***Quỳnh***: Dự án giúp nhận ra Khoa học dữ liệu là một quá trình "Thử nghiệm - Sai - Sửa" liên tục. Không có giải pháp nào hoàn hảo ngay từ bước đầu tiên, mà cần sự kiên nhẫn tinh chỉnh từng bước nhỏ, từ cách đặt câu hỏi định hướng cho đến cách trình bày kết quả cuối cùng.
