- **Tác động của Rào cản hành trình:** Định lượng được việc tỷ lệ thoát (Exit Rates) tại các trang quan trọng làm giảm xác suất dự báo mua hàng bao nhiêu % đối với từng nhóm khách tiềm năng.

Quy trình tiền xử lý bao gồm các bước chính sau:

**Bước 1: Tạo biến Tốc độ tương tác (Time per Page):**
* `Product_Speed` = `ProductRelated_Duration`/ `ProductRelated.`
* Vì một người xem 20 trang trong 2 phút (6 giây/trang) rất khác với người xem 5 trang trong 10 phút (120 giây/trang). Biến này giúp mô hình nhận diện nhóm Window Shoppers (lướt nhanh, không đọc kỹ) và nhóm Goal-Oriented (nghiên cứu sâu từng sản phẩm).

**Bước 2: Tạo biến tương tác hành vi (Interaction Feature):**
* `Value_Exit_Interaction`=  `Page Value` * (1 - `Exit Rate`)
* Giúp mô hình thấy mối quan hệ giữa giá trị trang và tỉ lệ thoát, ưu tiên những người đang ở page value cao và có ý định ở lại.

**Bước 3: Lọc đặc trưng (Feature Selection)**
* Các cột như `OperatingSystems`, `Browser`, `Region`, `TrafficType` phản ánh môi trường kỹ thuật hơn là tâm lý mua hàng. Để tập trung vào **ý định mua hàng**, ta chỉ giữ lại các cột mô tả hành vi trực tiếp và ngữ cảnh thời gian.
* Sau khi đã có `Time per Page` và `Duration`, các biến đếm số trang (`ProductRelated`, `Administrative`, `Informational`) trở nên dư thừa và gây ra hiện tượng đa cộng tuyến (như đã phân tích tương quan trước đó). Việc loại bỏ chúng giúp mô hình đơn giản hơn, tránh hiện tượng đa cộng tuyến.
  
**Bước 4: Biến đổi Log (Log transformation)**
* Áp dụng **$log(x+1)$** cho các toàn bộ các cột `Duration` (thời gian tương tác) và cho cột `Time per Page`.
* **Lý do:**
  * Các biến thời gian bị lệch phải cực đại do sự xuất hiện của các ngoại lai. Log Transformation giúp nén dải giá trị cực đoan, làm phân phối bớt nhọn và gần với phân phối chuẩn hơn, giúp các thuật toán không bị sai lệch bởi các giá trị cực đoan.

**Bước 5: Xử lý tính chất đa đỉnh(Binning)**
* **Exit Rate:** Chia thành 2 nhóm `Healthy` và `Critical` (Exit Rate = 0.2) để mô hình nhận diện rõ rào cản kỹ thuật.
    * **Lý do:** Dữ liệu `Exit Rate` có đặc tính hai đỉnh (Bimodal). Việc rời rạc hóa giúp mô hình nhận diện rõ ranh giới giữa một phiên truy cập bình thường và một phiên bị đứt gãy do lỗi kỹ thuật hoặc rào cản nội dung tại ngưỡng 0.20.
* **Special Day:** Gom nhóm 0.0 thành `Normal`, các giá trị < 0.8 thành `Near Holiday` và >= 0.8 thành `Peak Day`.
    * **Lý do:** Giải quyết vấn đề Lạm phát số 0. Việc chuyển sang dạng Categorical giúp mô hình đánh giá tác động của ngày lễ theo từng cấp độ tâm lý mua sắm thay vì giá trị số thuần túy.

**Bước 6: Chia dữ liệu**
* Chia tập **Train/Test** theo tỷ lệ **80/20**.
* **Lý do:**
    * **Split trước:** Để tránh rò rỉ thông tin từ tập Test vào tập Train.
  
**Bước 7: Mã hóa biến phân loại (Encoding) và Chuẩn hóa (Scaling)****
* **One-hot Encoding:** `Month`, `VisitorType`, `Exit_Status` và `SpecialDay_Group`.
    * **Lý do:** Logistic Regression yêu cầu dữ liệu số nhưng không được giả định thứ tự. One-hot Encoding giúp bắt được trọng số độc lập của từng giai đoạn lễ hội, giải quyết tính phi quy luật của các ngày đặc biệt.
* Sử dụng **RobustScaler** cho các biến số.
* **Lý do:**
  * **RobustScaler:** Sử dụng Trung vị (Median) và IQR thay vì Trung bình, giúp chuẩn hóa thang đo về cùng mức $[0, 1]$ mà không bị ảnh hưởng bởi các giá trị cực lớn vẫn còn tồn tại sau khi Log. 


```python

# --- BƯỚC 1: Tốc độ tương tác (Time per Page) ---
data['Product_Speed'] = data['ProductRelated_Duration'] / data['ProductRelated'].replace(0, 1)

# --- BƯỚC 2: Tương tác hành vi (Interaction Feature) ---
data['Value_Exit_Interaction'] = data['PageValues'] * (1 - data['ExitRates'])

# --- BƯỚC 3: Lọc đặc trưng (Feature Selection) ---
keep_cols = [
    'Administrative_Duration', 'ProductRelated_Duration', 'Product_Speed',
    'ExitRates', 'PageValues', 'Value_Exit_Interaction', 'SpecialDay',
    'Month', 'VisitorType', 'Weekend', 'Revenue'
]
data_ml = data[keep_cols].copy()

# --- BƯỚC 4: Biến đổi Log (Log transformation) ---
for col in ['Administrative_Duration', 'ProductRelated_Duration', 'Product_Speed']:
    data_ml[f'Log_{col}'] = np.log1p(data_ml[col])
data_ml = data_ml.drop(['Administrative_Duration', 'ProductRelated_Duration', 'Product_Speed'], axis=1)

# --- BƯỚC 5: Xử lý tính chất đa đỉnh (Binning) ---
data_ml['Exit_Status'] = pd.cut(data_ml['ExitRates'], 
                                bins=[-0.1, 0.19, 0.21], 
                                labels=['Healthy', 'Critical'])

def bin_special_day(val):
    if val == 0: return 'Normal'
    if val < 0.8: return 'Near_Holiday'
    return 'Peak_Day'
data_ml['SpecialDay_Group'] = data_ml['SpecialDay'].apply(bin_special_day)

data_ml['Weekend'] = data_ml['Weekend'].astype(int)
data_ml['Revenue'] = data_ml['Revenue'].astype(int)


# --- BƯỚC 6: Chia dữ liệu (Split) ---
# Chia trước khi Encoding/Scaling để tránh Data Leakage
X = data_ml.drop('Revenue', axis = 1)
y = data_ml['Revenue']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state = 42)

# --- BƯỚC 7.1: One-Hot Encoding  ---
def encode(train_df, test_df, columns):
    # Kết hợp để đảm bảo cấu trúc cột giữa Train và Test là đồng nhất
    combined = pd.concat([train_df, test_df], axis=0)
    # drop_first = True để tránh "Dummy Variable Trap" (Đa cộng tuyến hoàn hảo)
    combined_enc = pd.get_dummies(combined, columns=columns, drop_first=True)
    
    n_train = len(train_df)
    return combined_enc.iloc[:n_train], combined_enc.iloc[n_train:]

cat_cols = ['Month', 'VisitorType', 'Exit_Status', 'SpecialDay_Group']
X_train_final, X_test_final = encode(X_train, X_test, cat_cols)

# --- BƯỚC 7.2: Chuẩn hóa thang đo (Scaling) ---
num_cols = ['ExitRates', 'PageValues', 'Value_Exit_Interaction', 'SpecialDay', 
            'Log_Administrative_Duration', 'Log_ProductRelated_Duration', 'Log_Product_Speed']

scaler = RobustScaler()
X_train_final[num_cols] = scaler.fit_transform(X_train_final[num_cols])
X_test_final[num_cols] = scaler.transform(X_test_final[num_cols])

```

```python
# Chuyển về dạng NumPy array để nạp vào model Scratch

# 1. Đảm bảo toàn bộ bảng dữ liệu là kiểu số thực
X_train_final = X_train_final.astype(float)
X_test_final = X_test_final.astype(float)

# 2. Chuyển sang NumPy Array
X_train_arr = X_train_final.values
X_test_arr = X_test_final.values
y_train_arr = y_train.values.astype(float)
y_test_arr = y_test.values.astype(float)

# Kiểm tra
print(X_train_arr.dtype)
```

## 1. Mô hình Logistic Regression (Hồi quy Logistic)

### 1.1. Cơ sở toán học
Mô hình Logistic Regression dự báo xác suất $P(y=1|X)$ thông qua việc kết hợp một hàm tuyến tính với hàm kích hoạt phi tuyến (**Sigmoid**).

* **Hàm tuyến tính:** $$z = w_1x_1 + w_2x_2 + ... + w_nx_n + b = Xw + b$$

* **Hàm Sigmoid:** Giúp ép giá trị $z$ (từ $-\infty$ đến $+\infty$) về khoảng $[0, 1]$.
$$\sigma(z) = \frac{1}{1 + e^{-z}}$$

* **Hàm mất mát (Loss Function - Binary Cross Entropy):** Đo lường sự sai khác giữa xác suất dự báo $\hat{y}$ và thực tế $y$.
$$J(w, b) = -\frac{1}{m} \sum_{i=1}^{m} [y^{(i)} \log(\hat{y}^{(i)}) + (1 - y^{(i)}) \log(1 - \hat{y}^{(i)})]$$

---

### 1.2. Các bước huấn luyện (Gradient Descent)

1. **Khởi tạo:** Gán các trọng số $w$ ban đầu bằng 0 hoặc ngẫu nhiên.
2. **Dự báo (Forward Pass):** Tính $z$ và sau đó tính $\hat{y} = \sigma(z)$.
3. **Tính đạo hàm (Gradients):**
    * $\frac{\partial J}{\partial w} = \frac{1}{m} X^T(\hat{y} - y)$
    * $\frac{\partial J}{\partial b} = \frac{1}{m} \sum(\hat{y} - y)$
4. **Cập nhật (Update):** $w = w - \alpha \frac{\partial J}{\partial w}$ (với $\alpha$ là tốc độ học - learning rate).
5. **Lặp lại:** Cho đến khi hàm Loss hội tụ (đạt cực tiểu).

```python
class LogisticRegression:
    def __init__(self, lr=0.1, epochs=2000):
        self.lr = lr          # Tốc độ học (Learning Rate) - Kiểm soát mức độ cập nhật trọng số mỗi bước
        self.epochs = epochs  # Số lần lặp - Đảm bảo thuật toán có đủ thời gian để hội tụ về điểm tối ưu
        self.weights = None   # Hệ số beta (w) - Đại diện cho mức độ ảnh hưởng của từng biến
        self.bias = None      # Hệ số chặn (b) - Điểm xuất phát của mô hình khi các biến bằng 0

    def _sigmoid(self, z):
        z = np.asarray(z, dtype=float)
        return 1.0 / (1.0 + np.exp(-np.clip(z, -500, 500))) # clip để tránh lỗi tràn số exp

    def fit(self, X, y):
        # Bước 1: Khởi tạo tham số: Ban đầu chưa biết biến nào quan trọng,  gán tất cả bằng 0
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Bước 2: Vòng lặp Gradient Descent 
        for i in range(self.epochs):
            # TÍNH TOÁN DỰ BÁO (Forward Pass)
            model = np.dot(X, self.weights) + self.bias
            y_predicted = self._sigmoid(model)

            # TÍNH TOÁN SAI SỐ VÀ ĐẠO HÀM (Backpropagation)
            dw = (1.0 / n_samples) * np.dot(X.T, (y_predicted - y))
            db = (1.0 / n_samples) * np.sum(y_predicted - y)

            # CẬP NHẬT TRỌNG SỐ
            self.weights -= self.lr * dw
            self.bias -= self.lr * db

    def predict_proba(self, X):
        # Xác suất để vẽ đường cong ROC và tính AUC
        model = np.dot(X, self.weights) + self.bias
        return self._sigmoid(model)

    def predict(self, X, threshold=0.5):
        # Chuyển xác suất thành quyết định nhị phân để lập ma trận nhầm lẫn
        return np.array([1.0 if i > threshold else 0.0 for i in self.predict_proba(X)])
```

```python
# --- HUẤN LUYỆN LOGISTIC REGRESSION ---
print("Đang huấn luyện Logistic Regression...")
lr_scratch = LogisticRegression(lr=0.1, epochs=3000)
lr_scratch.fit(X_train_arr, y_train_arr)

print("Huấn luyện xong Logistic Regression...")
```

## 2. Mô hình Random Forest (Rừng ngẫu nhiên)

### 2.1. Cơ sở toán học
Random Forest dựa trên sự kết hợp của nhiều **Decision Tree** (Cây quyết định) thông qua kỹ thuật **Bagging** (Bootstrap Aggregating).

* **Chỉ số Gini (Gini Impurity):** Dùng để đánh giá độ "tinh khiết" của một nút sau khi chia. Điểm chia tối ưu là điểm làm giảm Gini nhiều nhất.
    $$Gini(D) = 1 - \sum_{i=1}^{k} p_i^2$$
    *(Trong đó $p_i$ là tỷ lệ của lớp $i$ trong tập dữ liệu $D$)*



* **Kỹ thuật Bagging:**
    * **Bootstrap:** Lấy mẫu ngẫu nhiên có lặp lại từ tập huấn luyện gốc để tạo ra các tập dữ liệu con (subsets).
    * **Feature Randomness:** Tại mỗi nút, chỉ chọn ngẫu nhiên một nhóm nhỏ các biến (features) để tìm điểm cắt, giúp giảm sự tương quan giữa các cây và chống Overfitting.

### 2.2. Các bước huấn luyện
1.  **Lấy mẫu:** Tạo ra $n$ tập dữ liệu con bằng phương pháp Bootstrap.
2.  **Xây dựng cây:** Với mỗi tập con, xây dựng một cây quyết định hoàn chỉnh:
    * Chọn ngẫu nhiên $k$ biến từ bộ dữ liệu.
    * Sử dụng chỉ số Gini để chọn biến và ngưỡng cắt (threshold) tối ưu.
    * Phát triển cây đến độ sâu tối đa (`max_depth`).
3.  **Tập hợp kết quả (Aggregating):**
    * Đối với bài toán phân loại, kết quả cuối cùng là sự biểu quyết số đông (**Majority Voting**) từ tất cả các cây trong rừng.

```python
class DecisionTree:
    def __init__(self, max_depth=10, n_features=None):
        self.max_depth = max_depth
        self.tree = None
        self.n_features = n_features
        # Khởi tạo mảng lưu trữ tầm quan trọng
        self.feature_importances = None 

    def _gini(self, y):
        m = len(y)
        if m == 0: return 0
        p = np.sum(y == 1) / m
        return 1.0 - p**2 - (1-p)**2

    def _build_tree(self, X, y, depth=0):
        n_samples, n_features = X.shape
        
        # Khởi tạo feature_importances nếu là nút gốc (depth=0)
        if depth == 0:
            self.feature_importances = np.zeros(n_features)

        # ĐIỀU KIỆN DỪNG
        if depth >= self.max_depth or len(np.unique(y)) == 1:
            return np.mean(y)

        best_gini_gain_weighted = -1 # Để tìm mức giảm Gini lớn nhất
        best_split = None
        parent_gini = self._gini(y)

        # TÌM ĐIỂM CHIA TỐI ƯU
        for feat_idx in range(n_features):
            thresholds = np.unique(X[:, feat_idx])
            for thr in thresholds:
                left_idx = np.where(X[:, feat_idx] <= thr)[0]
                right_idx = np.where(X[:, feat_idx] > thr)[0]
                
                if len(left_idx) == 0 or len(right_idx) == 0: continue
                
                # Gini trung bình có trọng số sau khi chia
                gini_left = self._gini(y[left_idx])
                gini_right = self._gini(y[right_idx])
                current_gini = (len(left_idx) * gini_left + len(right_idx) * gini_right) / n_samples
                
                # Tính Gini Gain 
                gini_gain = parent_gini - current_gini
                
                if gini_gain > best_gini_gain_weighted:
                    best_gini_gain_weighted = gini_gain
                    best_split = (feat_idx, thr, left_idx, right_idx, gini_gain)

        # XÂY DỰNG NHÁNH CON 
        if best_split:
            idx, thr, left, right, gain = best_split
            
            # CẬP NHẬT IMPORTANCE: Trọng số hóa theo số mẫu tại nút
            # Công thức: (n_samples_tại_nút / tổng_mẫu) * gini_gain
            self.feature_importances[idx] += n_samples * gain
            
            return {
                'feature': idx, 'threshold': thr,
                'left': self._build_tree(X[left], y[left], depth + 1),
                'right': self._build_tree(X[right], y[right], depth + 1)
            }
            
        return np.mean(y)

    def fit(self, X, y):
        self.tree = self._build_tree(X, y)
        # Chuẩn hóa tầm quan trọng về khoảng [0, 1] sau khi fit xong
        if self.feature_importances is not None:
            total = np.sum(self.feature_importances)
            if total > 0:
                self.feature_importances /= total

    def _predict_one(self, x, tree):
        if not isinstance(tree, dict): return tree
        if x[tree['feature']] <= tree['threshold']:
            return self._predict_one(x, tree['left'])
        return self._predict_one(x, tree['right'])

    def predict(self, X):
        return np.array([self._predict_one(x, self.tree) for x in X])
```

```python
class RandomForest:
    def __init__(self, n_trees=10, max_depth=10):
        self.n_trees = n_trees
        self.max_depth = max_depth # Lưu max_depth để truyền vào cây con
        self.trees = []
        self.feature_importances_ = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        # Khởi tạo mảng importance bằng 0 với số lượng đặc trưng tương ứng
        self.feature_importances_ = np.zeros(n_features)
        
        for i in range(self.n_trees):
            # Bootstrap: Lấy mẫu ngẫu nhiên có lặp lại
            indices = np.random.choice(n_samples, n_samples, replace=True)
            
            # Tạo cây mới với max_depth và n_features đã biết
            tree = DecisionTree(max_depth=self.max_depth, n_features=n_features)
            tree.fit(X[indices], y[indices])
            
            self.trees.append(tree)
            
            # Cộng dồn importance từ từng cây vào rừng
            if tree.feature_importances is not None:
                self.feature_importances_ += tree.feature_importances
        
        # CHIA TRUNG BÌNH: Lấy mức độ quan trọng trung bình của cả rừng
        self.feature_importances_ /= self.n_trees

    def predict_proba(self, X):
        # Lấy trung bình xác suất từ tất cả các cây (Soft Voting)
        tree_preds = np.array([tree.predict(X) for tree in self.trees])
        return np.mean(tree_preds, axis=0)
```

```python
# --- HUẤN LUYỆN RANDOM FOREST ---
print("Đang huấn luyện Random Forest...")
rf_scratch = RandomForest(n_trees=15, max_depth=8)
rf_scratch.fit(X_train_arr, y_train_arr)

print("Huấn luyện xong Random Forest")
```

### 1. Classification Report

```python
def manual_classification_report(y_true, y_prob, threshold=0.5):
    # Bước 1: Ép kiểu về mảng 1 chiều và kiểu số nguyên (0, 1) Để đảm bảo phép so sánh '==' và '&' hoạt động chính xác trên từng phần tử
    y_true = np.asarray(y_true).flatten().astype(int)
    y_pred = (np.asarray(y_prob).flatten() > threshold).astype(int)
    
    # Bước 2: Tính toán các thành phần ma trận nhầm lẫn
    tp = np.sum((y_true == 1) & (y_pred == 1))
    tn = np.sum((y_true == 0) & (y_pred == 0))
    fp = np.sum((y_true == 0) & (y_pred == 1))
    fn = np.sum((y_true == 1) & (y_pred == 0))
    
    # Bước 3: Tính toán các chỉ số
    accuracy = (tp + tn) / len(y_true)
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    
    print("-" * 35)
    print("CLASSIFICATION REPORT")
    print("-" * 35)
    print(f"Accuracy:  {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall:    {recall:.4f}")
    print(f"F1-Score:  {f1:.4f}")
    print("-" * 35)
    print(f"TN: {tn} | FP: {fp}")
    print(f"FN: {fn} | TP: {tp}")
    print("-" * 35)
```

```python
print("--- LOGISTIC REGRESSION ---")
manual_classification_report(y_test_arr, y_prob_lr)

print("\n--- RANDOM FOREST ---")
manual_classification_report(y_test_arr, y_prob_rf)
```

2. **Sự đánh đổi (Trade-off):** Đối với lĩnh vực thương mại điện tử, việc bắt được đúng khách hàng có nhu cầu mua (**Recall cao**) thường mang lại giá trị kinh tế lớn hơn việc gửi nhầm quảng cáo cho một ít người không có nhu cầu.

---

#### Kết luận
**Random Forest là mô hình thực tế và hiệu quả hơn** để triển khai các chiến dịch can thiệp thời gian thực hoặc quảng cáo bám đuổi (retargeting) nhờ khả năng phân loại khách hàng mục tiêu vượt trội.

#### 1. Nhận xét về các đặc trưng thúc đẩy mua hàng (Trọng số dương - Màu xanh)

* **Tháng 11 (Month_Nov):** Đây là đặc trưng có trọng số dương cao nhất (~0.85). Yếu tố thời điểm cuối năm với các sự kiện mua sắm lớn (như Black Friday) là động cơ mạnh mẽ nhất thúc đẩy hành vi mua hàng.
* **Thời gian xem sản phẩm (Log_ProductRelated_Duration):** Trọng số dương đáng kể (~0.35). Minh chứng rằng khách hàng càng dành nhiều thời gian nghiên cứu thông tin sản phẩm, khả năng mua hàng càng cao.
* **Biến tương tác tự tạo (Value_Exit_Interaction):** Trọng số dương (~0.2) khẳng định hiệu quả của việc kết hợp giữa giá trị trang và việc giữ chân khách hàng.
* **Cuối tuần (Weekend):** Có tác động tích cực nhẹ, cho thấy khách hàng có xu hướng mua sắm thoải mái hơn vào những ngày nghỉ.

#### 2. Nhận xét về các rào cản mua hàng (Trọng số âm - Màu đỏ)
Các biến này đại diện cho những yếu tố ngăn cản hoặc làm giảm khả năng hoàn tất giao dịch.

* **Tháng 5 (Month_May):** Rào cản lớn nhất với trọng số âm gần -1.0. Đây là giai đoạn thấp điểm của mua sắm trực tuyến trong tập dữ liệu này.
* **Khách hàng quay lại (VisitorType_Returning_Visitor):** Trọng số âm mạnh (~ -0.8). Một phát hiện bất ngờ: khách hàng cũ có thể đang quay lại để so sánh giá hoặc kiểm tra thông tin, dẫn đến tỷ lệ chốt đơn thấp hơn so với khách hàng mới.
* **Tỷ lệ thoát (ExitRates):** Trọng số âm rõ rệt (~ -0.5). Tỷ lệ thoát trang cao là "kẻ thù" của doanh thu; khách hàng càng dễ rời đi thì cơ hội bán hàng càng thấp.
* **Áp lực ngày lễ (SpecialDay):** Tác động âm (~ -0.45). Nếu khách ghé thăm quá sát ngày lễ, họ có thể lo ngại về thời gian vận chuyển hoặc hết hàng, dẫn đến trì hoãn mua sắm.

---

#### 3. Insight chiến lược

* **Tối ưu hóa "Ngưỡng tương tác":**
    *  Vì `Log_ProductRelated_Duration` có trọng số dương mạnh, doanh nghiệp nên tập trung làm giàu nội dung trang sản phẩm để giữ chân khách hàng lâu hơn.

* **Giải pháp cho rào cản Exit Rates**:
    * Trọng số âm của `ExitRates` cho thấy cần can thiệp ngay lập tức. Nếu hệ thống nhận diện khách hàng ở trang có `PageValues` cao nhưng có dấu hiệu sắp thoát trang, hãy tung ra **Real-time Incentives** để đảo ngược tác động âm này.

* **Cá nhân hóa theo thời điểm**
    * **Tháng 11:** Dồn tổng lực ngân sách Marketing.
    * **Tháng 5:** Cần các chương trình kích cầu mạnh mẽ hơn để vượt qua rào cản mua sắm tự nhiên.

* **Phân tách khách hàng**:
    * Trọng số âm của `Returning Visitor` nhắc nhở doanh nghiệp không được chủ quan với khách cũ. Cần có các chính sách **"Tri ân khách hàng thân thiết"** riêng biệt để biến họ từ người tham khảo thành người mua hàng thực thụ.

#### 1.1 Các vấn đề cốt lõi

Dựa trên kết quả thực nghiệm từ các mô hình, ta có thể làm rõ các vấn đề cốt lõi sau:

* **Ngưỡng quyết định (The Tipping Point):** Phân tích cho thấy `PageValues` là yếu tố dự báo mạnh mẽ nhất (Gini Importance ~0.35). Kết hợp với sự vượt trội của Random Forest (**AUC = 0.9285**), chúng ta xác nhận rằng ý định mua hàng đạt mức "bùng phát" khi khách hàng tương tác với các trang có giá trị cao, thay vì chỉ tích lũy thời gian xem trang một cách tuyến tính đơn thuần.
  
* **Sự đánh đổi giữa Thời gian và Hiệu quả:** `Log_ProductRelated_Duration` có trọng số dương đáng kể (0.35), chứng minh thời gian xem trang lâu (*High Duration*) trong dữ liệu này là dấu hiệu của sự quan tâm sâu (**Goal-Oriented**) thay vì sự lưỡng lự. Tuy nhiên, biến tự tạo `Value_Exit_Interaction` đứng vị trí thứ 2 (~0.31) cho thấy thời gian chỉ thực sự có ý nghĩa nếu nó đi kèm với việc không rời trang sớm.

  
* **Định lượng rào cản:** `ExitRates` có trọng số âm rõ rệt (~ -0.5), đóng vai trò là "kẻ thù" lớn nhất của doanh thu trong toàn bộ hành trình khách hàng.

1.  **Chuyển dịch chiến lược:** Doanh nghiệp nên chuyển từ "giữ chân khách hàng thụ động" sang **"điều hướng hiệu quả"** bằng cách dẫn dắt người dùng tới các trang có `PageValues` cao thông qua hệ thống gợi ý.
2.  **Tối ưu ngân sách:** Sử dụng Random Forest cho phép tiếp cận thêm **~200 khách hàng tiềm năng** mà Logistic Regression đã bỏ lỡ, giúp tối ưu hóa hiệu quả của các chiến dịch quảng cáo bám đuổi (*Retargeting*).

#### 2. Phát hiện bất ngờ & Hạn chế

##### Bất ngờ
* **Khách hàng quay lại (`Returning_Visitor`):** Có trọng số âm mạnh (~ -0.8). Điều này đập tan giả định thông thường rằng khách cũ dễ mua hàng hơn. Thực tế, trong tập dữ liệu này, họ là nhóm đang lưỡng lự hoặc đang thực hiện so sánh giá nhiều hơn là có ý định chốt đơn ngay.

#####  Hạn chế:
* **Sai lầm loại I (False Positive):** Mô hình Random Forest vẫn nhầm lẫn khoảng **98-101 trường hợp** "người xem dạo" (*Window Shoppers*) thành người mua.
* **Hướng cải thiện:** Cần bổ sung thêm các biến về nguồn truy cập (*Traffic Source*) hoặc loại thiết bị để phân loại sâu hơn các nhóm "chỉ tham khảo" này.
